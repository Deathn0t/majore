```python
import os
import numpy as np
import re
from matplotlib import pyplot as plt


%matplotlib inline
```

```python
losses = []
with open('trace_loss', 'r', encoding='utf-8') as f:
    for line in f:
        if 'tensor' not in line:
            continue
        loss = float(line.split('tensor(')[1].split(',')[0])
        losses.append(loss)
```

```python
plt.plot(losses)
plt.yscale('log')
plt.show()
```
